```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/newduplex.tsv", sep='\t')
display(data)
```

```python
data['Starting_Strand_Length'] = data['end1'] - data['start1']
```

```python
print(data['Starting_Strand_Length'].describe()['std'])
print(data.groupby('Starting_Strand_Length').size())

plotdata = data.groupby([data['Starting_Strand_Length']]).size()
plotdata.plot(kind='bar', cmap='autumn', figsize=(7,7),legend=None)
```

```python
data.groupby('Starting_Strand_Length').agg(duplex = ('duplex', 'std'))
```

```python
data.groupby(['duplex', 'Starting_Strand_Length']).size().plot()
```

```python
data.groupby(['duplex', 'Starting_Strand_Length']).size().reset_index().to_csv('dataread.csv')
data.groupby(['duplex'])['Starting_Strand_Length'].value_counts().to_csv('dataread2.csv')
data.value_counts(["duplex", "Starting_Strand_Length"]).sort_values(ascending=False)
```
